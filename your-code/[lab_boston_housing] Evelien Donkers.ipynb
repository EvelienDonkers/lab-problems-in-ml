```python
def explore_data(df):
    # this function applies many exploratory techniques to a given dataframe
    display("Head",
            df.head()
            .style
            .background_gradient(cmap='Blues')
           )
    display("Data Types",
            df.dtypes
            .to_frame()
           )
    display("Data Types Count",
            df.dtypes
            .to_frame()[0]
            .value_counts()
            .to_frame()
           )
    display("Nan Values per Column",
            df.isna()
            .sum()
            .to_frame()
            .sort_values(by=[0], ascending=False)
           )
    display("Descriptive Statistics",
            df.describe()
           )
    display("Correlation Matrix",
            df.corr()
            .style.background_gradient(cmap='coolwarm')
            .set_precision(2)
           )
```

```python
regr2 = RandomForestRegressor(max_depth=2, random_state=0)
regr4 = RandomForestRegressor(max_depth=4, random_state=0)
regr6 = RandomForestRegressor(max_depth=6, random_state=0)
regr8 = RandomForestRegressor(max_depth=8, random_state=0)
regr10 = RandomForestRegressor(max_depth=10, random_state=0)
reg = [regr2,regr4,regr6,regr8,regr10]
```

```python
test = []
train = []
parameters = [2, 4, 6, 8 ,10]
for i in range(len(reg)):
    model = reg[i].fit(X_train, y_train)
    print(f'Max depth: {parameters[i]},  Score test: {reg[i].score(X_test, y_test)},  Score train: {reg[i].score(X_train, y_train)}')
    test.append(reg[i].score(X_test, y_test))
    train.append(reg[i].score(X_train, y_train))
```

Now, plot the score for each tree on the training set and on the testing set.

```python
sns.scatterplot(x = parameters, y = train, color = 'red')
sns.scatterplot(x = parameters, y = test, color = 'blue')
```
